```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page1 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&count=100&start=1&ref_=adv_nxt')
soup1 = BeautifulSoup(page1.content)

name = []

for i in soup1.find_all(class_="lister-item-header"):
    name.append(i.text.splitlines()[2])

rating = []
for i in soup1.find_all(class_="ratings-bar"):
    rating.append(i.text.split()[13])

release = []
for i in soup1.find_all(class_="lister-item-year text-muted unbold"):
    release.append(i.text)


print('*'*63)
print('\t\t\t IMDB’s Top rated 100 movies')
print('*'*63)

df_IMDB = pd.DataFrame({'Name of movie':name,'Rating of movie':rating,'Year of release of movie':release})

df_IMDB.head(100)
```

```python
page2 = requests.get('https://www.imdb.com/list/ls009997493/?sort=user_rating,desc&st_dt=&mode=detail&page=1')
soup2 = BeautifulSoup(page2.content)

name1=[]
for i in soup2.find_all('h3', class_="lister-item-header"):
    cells = i.find_all('a')
    name1.append(cells[0].text.strip())
    
rating1=[]
for i in soup2.find_all('div', class_="ipl-rating-star small"):
    rating1.append(i.text.replace('\n',''))

release1=[]
for i in soup2.find_all('span', class_="lister-item-year text-muted unbold"):
    release1.append(i.text)


print('*'*100)
print('\t\t IMDB’s Top rated 100 Indian movies')
print('*'*100)

df_Indian = pd.DataFrame({'Name of movie':name1,'Rating of movie':rating1,'Year of release of movie':release1})

df_Indian
```

```python
page3 = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup3 = BeautifulSoup(page3.content)

PresidentName=[]
for i in soup3.find_all('div', class_="presidentListing"):
    cells = i.find_all('h3')
    PresidentName.append(cells[0].text.strip())

Term=[]
for i in soup3.find_all('div', class_="presidentListing"):
    cells = i.find_all('p')
    Term.append(cells[0].text.strip().replace('Term of Office:',''))

print('*'*100)
print('\t\t\t Former Presidents List')
print('*'*100)

df_president=pd.DataFrame({'Name of President':PresidentName,'Term of Office':Term})
df_president
```

```python
page4_b= requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup4_b= BeautifulSoup(page4_b.content)

Player=[]
Team=[]
Rating=[]

for i in soup4_b.find_all('tr')[:11]:
    cells = i.find_all('td')
    if len(cells) == 5:
        Player.append(cells[1].text.strip().replace('\n'," , "))
        Team.append(cells[2].text.strip())
        Rating.append(cells[3].text.strip())
        
print('*'*100)
print(" Top 10 ODI Batsmen in Men's Cricket")
print('*'*100)

df_icc_batting= pd.DataFrame({'Player':Player,'Team':Team,'Rating':Rating})  
df_icc_batting
```

```python
page5 = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
soup5 = BeautifulSoup(page5.content)

Matches = []
Points = []
Rating = []
Team = []
Ranking = []

for i in soup5.find_all('tr')[:11]:
    cells = i.find_all('td')
    if len(cells) == 5:
        Ranking.append(cells[0].text.strip())
        Team.append(cells[1].text.strip().replace('\n'," , "))
        Matches.append(cells[2].text.strip())
        Points.append(cells[3].text.strip())
        Rating.append(cells[4].text.strip())

print('*'*100)
print("     Top 10 ODI Teams in Women's Cricket")
print('*'*100)

df_Womens = pd.DataFrame({' Ranking':Ranking,'Team':Team,'Matches':Matches,'Points':Points,'Rating':Rating})  
df_Womens
```

```python
page6= requests.get('https://www.cnbc.com/world/?region=world')
soup6= BeautifulSoup(page6.content)

Headline = []
for i in soup6.find_all('a', class_="LatestNews-headline"):
    Headline.append(i.get('title'))


Time = []
for i in soup6.find_all('span', class_="LatestNews-wrapper"):
    Time.append(i.text)

News_Link = []
for i in soup6.find_all('a',class_="LatestNews-headline"):
    News_Link.append(i.get('href'))
    
print('*'*100)
print("News Details are below")
print('*'*100)


News= pd.DataFrame({'Headline':Headline,'Time':Time,'News link':News_Link})  
News
```

```python
page8= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup8= BeautifulSoup(page8.content)

Paper_Title=[]
for i in soup8.find_all('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Paper_Title.append(i.text)

Authors=[]
for i in soup8.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)  

Published_Date=[]
for i in soup8.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
    Published_Date.append(i.text)  

Paper_URL=[]
for i in soup8.find_all('a', class_="sc-5smygv-0 fIXTHm"):
    Paper_URL.append(i.get('href'))

print('*'*125)
print('\t\t\t\t\t\t Most Downloaded Articles are here')
print('*'*125)

Articles=pd.DataFrame({'Paper Title': Paper_Title,'Authors': Authors,'Published Date': Published_Date, 'Paper URL':Paper_URL})
Articles
```

```python
page9= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup9= BeautifulSoup(page9.content)

restaurant_name= []
for i in soup9.find_all('a', class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)

Cuisine=[]
for i in soup9.find_all('div', class_="detail-info"):
    Cuisine.append((i.text.split()[6])+(' ')+(i.text.split()[7]))


location=[]
for i in soup9.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)

Ratings= []
for i in soup9.find_all('div', class_="restnt-rating rating-4"):
    Ratings.append(i.text)

Image_URL= []
for i in soup9.find_all('img', class_="no-img"):
    Image_URL.append(i['data-src'])


print('*'*100)
print('\t\t\t\t\t\t Restaurant Details are here')
print('*'*100)

dineout=pd.DataFrame({'Restaurant_Name':restaurant_name,'Cuisine':Cuisine,'Location':location,'Ratings':Ratings,'Image_URL':Image_URL})
dineout
```
